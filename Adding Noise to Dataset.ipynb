```python
def truncated_normal(mean, stddev, minval, maxval, count):
        return np.clip(np.random.normal(mean, stddev, count), minval, maxval)
```

```python
ages = truncated_normal(1980, 15, 1948, 2000, len(df2['GIID']))
ages = [int(n) for n in ages]

months = np.random.uniform(1,13,len(df2['GIID']))
months = [int(x) for x in months]

monthToYear = {1:31, 2:28, 3:31, 4:30, 5:31, 6:30, 7:31, 8:31, 9:30, 10:31, 11:30, 12:31}
days = [1] * len(months)

for i in range(len(months)):
    numdays = monthToYear[months[i]]
    if months[i]==2:
        numdays = np.random.choice([28,28,28,29])
    days[i] = np.random.randint(1, numdays+1)
    

df2['YO'] = ages
df2['MO'] = months
df2['DO'] = days
```

```python
qwertyKeyboard = ['qwertyuiop', 'asdfghjkl', 'zxcvbnm']
let = {}
for q in range(len(qwertyKeyboard)):
    for i in range(len(qwertyKeyboard[q])):
        tmp = []
        if i != 0:
            tmp.append(qwertyKeyboard[q][i-1])
        if i != len(qwertyKeyboard[q])-1:
            tmp.append(qwertyKeyboard[q][i+1])
        if q != len(qwertyKeyboard)-1:
            tmp.append(qwertyKeyboard[q+1][min(i, len(qwertyKeyboard[q+1])-1)])
        if q != 0:
            tmp.append(qwertyKeyboard[q-1][i])
        let[qwertyKeyboard[q][i]] = tmp
```

```python
new_firsts = [0]*len(dfupdate['INDEX'])
ind = 0
for i, j in dfupdate.iterrows():
    toChange = np.random.randint(0, min(2, len(j['PI_FIRST_NAME'])))
    newBeg = ''
    for q in range(toChange):
        if j['PI_FIRST_NAME'][q:min(q+1, len(j['PI_FIRST_NAME']))].lower() == ' ':
            break
        toPick = let[j['PI_FIRST_NAME'][q:min(q+1, len(j['PI_FIRST_NAME']))].lower()]
        newBeg += random.choice(toPick).upper()
    new_firsts[ind] = newBeg + j['PI_FIRST_NAME'][toChange::]
    ind += 1
```

```python
dfupdate['PI_FIRST_NAME'] = new_firsts
dfupdate['CHANGED_FIRST_NAME'] = True
df.update(dfupdate)
update_list_firsts = dfupdate.index.tolist() # from  cᴏʟᴅsᴘᴇᴇᴅ  :)
```
